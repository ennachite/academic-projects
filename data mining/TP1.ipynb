# KNN (K-nearest neighbors)

**2.1. Construire le modèle K-nearest neighbors**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
X = iris.data
Y = iris.target 
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X, Y)

y_model = model.predict(X)
accuracy_score(Y, y_model)
```

On constate qu'on a obtenu une précision de 100% mais cela ne veut pas dire que notre modele sera performant pour n'importe quel exemple mais plutôt que pour les records dont on dispose dans notre dataset et puisqu'ils sont ceux qu'il a utilisés pour s'entrainer.


**2.2.a) L'approche Holdout**

```python
#Hold out 
model2 = KNeighborsClassifier(n_neighbors=5)
X1, X2, Y1, Y2 = train_test_split(X, Y, random_state=np.random, train_size=0.7)

y_pred = model.fit(X1, Y1).predict(X2)

y2_model = model.predict(X2)

accuracy_score(Y2, y2_model)
```

En éxecutant le code plusieurs fois on obtient des valeurs de precision différentes ce qui est normale puisqu'à chaque fois
qu'on l'execute on genere un subDataSet qui est différent du precedant à l'aide de la fonction random de la bibliotheque
numpy.

**2.2.b) Cross-validation (Validation croisée)**

```python
#2 fold
X1, X2, Y1, Y2 = train_test_split(X, Y, random_state=0, train_size=0.5)

y2_pred = model.fit(X1, Y1).predict(X2)
y1_pred = model.fit(X2, Y2).predict(X1)

accuracy_score(Y1, y1_pred), accuracy_score(Y2, y2_pred)
```

**2.2.c) 5-folds cross-validation**

```python
#5 fold
from sklearn.model_selection import cross_val_score  
cross_val_score(model, X, Y, cv=5)
```

Elle est due au choix aléatoire des exemples appartenant à chaque Fold, vu qu'à une autre réexécution du code il se peut qu'on se trouve avec une nouvelle composition du SubDataSet d'entraînement ou de test.

**2.2.d) Leave one out cross validation**

```python
#lOOCV 
from sklearn.model_selection import LeaveOneOut  
scores = cross_val_score(model, X, Y, cv =LeaveOneOut() ) 
scores
```

```python
scores.mean()
```

On constate qu'on obtient toujours le même résultat, ce qui est expliqué par le fait qu'on a testé et qu'on s'est entraîné avec toutes les possibilités existantes.

```python
X1, X2, y1, y2 = train_test_split(X,Y, random_state= np.random, train_size=0.7) 
y_pred=model2.fit(X1,y1).predict(X2) 
y2_model=model2.predict(X2) 

 

cm1= confusion_matrix(y2, y2_model) 
print("Confusion Matrix: \n", cm1) 
print(accuracy_score(y2, y2_model)) 

 

print(precision_score(y2, y2_model, average='weighted')) 
print(recall_score(y2, y2_model, average='macro')) 

 

print("Accuracy:", accuracy_score(y2, y2_model)) 
print("Confusion matrix:", confusion_matrix(y2, y2_model)) 
print("Classification report :", classification_report(y2, y2_model)) 
```

*    **Hold out 70% 30%:**

```python
# Hold out 70%-30%
import numpy as np
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

iris = load_iris()
X = iris.data
y = iris.target
model = KNeighborsClassifier(n_neighbors=5)
X1, X2, y1, y2 = train_test_split(X, y, random_state= np.random, train_size=0.7) 
y2_pred = model.fit(X1, y1).predict(X2)
y2_model = model.predict(X2)

print("Accuracy: ", accuracy_score(y2, y2_model))
print("Precision: ", precision_score(y2, y2_model, average='weighted'))
print("Recall: " , recall_score(y2, y2_model, average='macro'))
print("F1 score: ", f1_score(y2, y2_model, average='macro'))
```

*    **5 Folds-cross-validation:**

```python
# 5 folds cross-validation
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

iris = load_iris()
X = iris.data
y = iris.target
model = KNeighborsClassifier(n_neighbors=5)
scores_acc = cross_val_score(model, X, y, cv=5)
scores_prec = cross_val_score(model, X, y, cv=5, scoring='precision_macro')
scores_rec = cross_val_score(model, X, y, cv=5, scoring='recall_macro')
scores_f1 = cross_val_score(model, X, y, cv=5, scoring='f1_macro')

print("Accuracy: ", scores_acc.mean())
print("Precision: ", scores_prec.mean())
print("Recall: " , scores_rec.mean())
print("F1 score: ", scores_f1.mean())
```

**2.4. Trouver les paramètres optimaux : Grid Search**

```python
#Grid search  
from sklearn.model_selection import GridSearchCV 
#create a new KNN model  
Knn2 = KNeighborsClassifier()  
K_range = list(range(1,31)) 
pram_grid = dict(n_neighbors = K_range) 
grid = GridSearchCV(Knn2, pram_grid, cv = 10, scoring = 'accuracy') 
grid.fit(X,y) 
print(grid.best_score_) 
print(grid.best_params_) 
print(grid.best_estimator_)
```
